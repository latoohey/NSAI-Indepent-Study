```python
from imodels import RuleFitClassifier, BayesianRuleListClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer # Use sklearn instead
import numpy as np

# 1. Load Data
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

# 2. Discretize for BRL (Converts numbers to "One-Hot" encoded binary flags)
est = KBinsDiscretizer(n_bins=3, encode='onehot-dense', strategy='quantile', subsample=None)
X_train_bin = est.fit_transform(X_train)
X_test_bin = est.transform(X_test)

# Create readable feature names for the bins
feature_names_bin = est.get_feature_names_out(data.feature_names)

# 3. Fit BRL
brl = BayesianRuleListClassifier()
brl.fit(X_train_bin, y_train, feature_names=list(feature_names_bin))

print("\n--- Bayesian Rule List Results ---")
print(brl)
```
